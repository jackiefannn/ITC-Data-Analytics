#### combine Christian's cleaning process

```python
pittsburgh_accident_df=pittsburgh_accident_df.drop(columns=[ 'sex_of_driver', 'age_of_driver','police_force', 'was_vehicle_left_hand_drive?', 
                           'local_authority_(district)', 'local_authority_(highway)','engine_capacity_(cc)'
                          ,'did_police_officer_attend_scene_of_accident','journey_purpose_of_driver','propulsion_code'])
```

```python
len(pittsburgh_accident_df.columns)
```

```python
pittsburgh_accident_df=pittsburgh_accident_df[(pittsburgh_accident_df.carriageway_hazards != 4) &
                       (pittsburgh_accident_df.carriageway_hazards != 5 )&
                       (pittsburgh_accident_df.carriageway_hazards != 7) &
                       (pittsburgh_accident_df.carriageway_hazards != -1)]
#firstDF = firstDF[pittsburgh_accident_df.carriageway_hazards != 5]
#firstDF = firstDF[pittsburgh_accident_df.carriageway_hazards != 7]
#firstDF = firstDF[pittsburgh_accident_df.carriageway_hazards != -1]
```

```python
pittsburgh_accident_df.carriageway_hazards.unique()
```

```python
pittsburgh_accident_df= pittsburgh_accident_df[(pittsburgh_accident_df.junction_control != -1) | (pittsburgh_accident_df.weather_conditions != -1) | (pittsburgh_accident_df.road_surface_conditions != -1) 
                  | (pittsburgh_accident_df.junction_detail != -1) | (pittsburgh_accident_df.road_type != -1) 
                  | (pittsburgh_accident_df.weather_conditions != -6) | (pittsburgh_accident_df.urban_or_rural_area != -1) | (pittsburgh_accident_df.light_conditions != -1) ]
```

### Sort by total

```python
pd.crosstab(index=pittsburgh_accident_df.year, 
            columns=pittsburgh_accident_df.accident_severity, 
            margins=True, 
            margins_name='Total').sort_values('Total',ascending=False)
```

```python
pd.crosstab(index=pittsburgh_accident_df.street_name, 
            columns=pittsburgh_accident_df.accident_severity, 
            margins=True, 
            margins_name='Total').sort_values('Total',ascending=False).head(20)
```

```python
pd.crosstab(index=[pittsburgh_accident_df.street_name,pittsburgh_accident_df.year], 
            columns=pittsburgh_accident_df.accident_severity, 
            margins=True, 
            margins_name='Total').sort_values('Total',ascending=False).head(20)
```

```python
pd.crosstab(index=pittsburgh_accident_df.street_name, 
            columns=pittsburgh_accident_df.accident_severity, 
            margins=True, 
            margins_name='Total').corr()
```
